#ALTERNATE METHOD

#SEPERATING FEATURES AND TARGET VARIABLES


```
Labels=train['Species']
train=train.drop('Species',axis=1)
Features=train[:]
```

```
np.array(Features).shape
```

```
np.array(Labels).shape
```

#SPLITTING DATA INTO TEST(30%) AND TRAIN(70%)

```
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(Features,Labels,test_size=0.3)
```

#IMPLIMENTING THE DECISIONTREECLASSIFIER

```
from sklearn.tree import DecisionTreeClassifier
dmodel=DecisionTreeClassifier()
dmodel.fit(xtrain,ytrain)
```

```
dmodel.score(xtest,ytest)
```

#VISUALIZING THE MODEL


```
from sklearn import tree
plt.figure(figsize = (10,10))
tree.plot_tree(dmodel,filled=True)
```

```

def predictions(Id,SepalLengthCm,SepalWidthCm,PetalLengthCm,PetalWidthCm):
  prediction=dmodel.predict([[Id,SepalLengthCm,SepalWidthCm,PetalLengthCm,PetalWidthCm]])
  if(prediction==([0])):
    print('Iris-setosa')
  elif(prediction==([1])):
    print('Iris-versicolor')
  elif(prediction==([2])):
    print('Iris-virginica')
  
```

#CHECKING OUTPUT ON RANDOM INPUTS
